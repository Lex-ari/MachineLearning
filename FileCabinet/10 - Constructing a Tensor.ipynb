# Tensors are the fundamental data structure in PyTorch
A **tensor** is a multidimensional array. The dimensionality of a tensor coincides with the number of indexes used to refer to scalar values within the tensor.

PyTorch tensors have the following important features,which are needed in order to implement a modern deep learning library!
- the ability to perform very fast operations on graphical processing units (GPUs),
- distribute operations on multiple devices or machines, and
- keep track of the graph of computations that created them.

## Import the torch module

```python
import torch
```

## Construct a tensor of specified size

```python
# create a one-dimensional tensor of size 3 filled with 1s
a = torch.ones(3)
print('a-->',a)
print()

# get the size of the tensor along each dimension.
print('a.shape-->',a)
print('a.dtype-->',a.dtype)
print()

# access an element using its zero-based index
print('a[0]-->',a[0])
print('a[1]-->',a[1])
print('a[2]-->',a[2])
print()

# assign a new value to a tensor element
a[1] = 2
print('a-->',a)
print('type(a[1])-->',type(a[1]))
print()

# convert a one-element tensor to Python float scalar
x = float(a[1])
print('x-->',x)
print('type(x)-->',type(x))
print()
```

```python
# create a 2D tensor of size 4 by 3 filled with  random numbers from a uniform distribution on the interval [0,1)
a = torch.rand(4,3)
print('a-->',a)
print()

# get the size of the tensor along each dimension.
print('a.shape-->',a)
print()

# create a 2D tensor of size 4 by 3 filled with 0s (floating-point numbers)
a = torch.zeros(4,3)
print('a-->',a)
print('a.dtype-->',a.dtype)
print()

# create a 2D tensor of size 4 by 3 filled with 0s (integer numbers)
a = torch.zeros(4,3, dtype = torch.long)
print('a-->',a)
print('a.dtype-->',a.dtype)
print()
```

```python
a = torch.ones(4,3, dtype = torch.long)
print('a-->',a)
print('a.dtype-->',a.dtype)
print()

# return a tensor with the same size as a that is filled with zeros
# torch.zeros_like(input) is equivalent to torch.zeros(input.size(), dtype=input.dtype, layout=input.layout, device=input.device).
b = torch.zeros_like(a)
print('b-->',b)
print('a.dtype-->',a.dtype)
print()

# return a tensor with the same size as a that is filled with zeros, but change the data type
c = torch.zeros_like(a, dtype = torch.float)
print('c-->',c)
print('c.dtype-->',c.dtype)
```

## Construct a tensor from a Python list (with given values)

```python
# create a 2D tensor by passing a Python list of lists to the PyTorch tensor constructor
a = torch.tensor([[4.0, 1.0],[5.0, 3.0], [2.0, 1.0]])
print('a-->',a)
print()

# get the size of the tensor along each dimension.
print('a.shape-->',a.shape)
```

```python
# Note: In each of the following, another tensor is created that presents a different view of the same underlying data.
# access an individual element in the tensor using two indices
b = a[0, 1]
print('b-->', b)
print()

# access the first point in the tensor
c = a[0]
print('c-->', c)
print()

# get all rows after the first; implicitly all columns
d = a[1:]
print('d-->', d)
print()

# get all rows after the first; all columns
e = a[1:, :]
print('e-->', e)
print()

# get all rows after the first; first column
f = a[1:, 0]
print('f-->', f)
print()

# adds a dimension of size 1, just like unsqueeze operation (we will go over later)
g = a[None]
print('g-->', g)
print('g.shape-->', g.shape)
```

# Tensor element types

The **dtype** argument specifies the possible values the tensor can hold (integers versus floatingpoint numbers) and the number of bytes per value

3.png

**Floating-point numbers** are the way a network deals with information. The machine learning/deep learning process begins by converting our input into floating-point numbers.

- Floating-point elements: the default data type for tensors is **32-bit floating-point**
- Integer elements: Indexing tensors is a **64-bit integer** data type, as tensors can be used as indexes in other tensors.
- Boolean elements: **bool** tensors usually is used to indicate whether each individual element satisfies a condition such as a > 1.0

4.png

## Specify the proper dtype

```python
# specify the proper dtype as an argument to the tensor constructor, to allocate a tensor of the right numeric type
double_a = torch.ones(10, 2, dtype = torch.double)
short_b = torch.tensor([[1,2],[3,4]], dtype = torch.short)

# find out the dtype for a tensor by accessing the corresponding attribute:
print('double_a.dtype -->', double_a.dtype)
print('short_b.dtype -->', short_b.dtype)
print()

# alternatively, we can cast the output of a tensor creation function to the right type using the corresponding casting method
double_c = torch.randn(10, 2).double()
short_d = torch.ones(10, 2).short()
print('double_c.dtype -->', double_c.dtype)
print('short_d.dtype -->', short_d.dtype)
print()

# or, the more convenient way to use "to" method:
double_e = torch.randn(10, 2).to(torch.double)
short_f = torch.ones(10, 2).to(dtype=torch.short)
print('double_e.dtype -->', double_e.dtype)
print('short_f.dtype -->', short_f.dtype)
```

## When mixing input types in operations

```python
# When mixing input types in operations, the inputs are converted to the larger type automatically.
double_a = torch.tensor([2.0], dtype = torch.double)
short_b = torch.tensor([3], dtype = torch.short)

c = double_a * short_b

print('double_a.dtype -->', double_a.dtype)
print('short_b.dtype -->', short_b.dtype)
print('c-->', c)
```

# Tensors vs Python Lists
- Memory spaces
>- Python lists are collections of Python objects that are individually allocated in memory.
>- PyTorch tensors are views over contiguous memory blocks containing unboxed C numeric types rather than Python objects.
- Tensors support arithmetic operations

## Memory space
Storing a 1D tensor of 1,000,000 float numbers will require exactly 4,000,000 contiguous bytes, plus a small overhead for the metadata (such as dimensions and numeric type).

## Arithmetic operation

```python
# add two Python lists
a_py = [1, 2]
b_py = [3, 4]

c_py = a_py + b_py
print('c_py-->',c_py)
```

```python
import torch

# add two Pytorch tensors
a = torch.tensor([1,2])
b = torch.tensor([3,4])

c = a + b
print('c-->',c)
```

```python
import numpy as np

# add two numpy arrays
a_np = np.array([1,2])
b_np = np.array([3,4])

c_np = a_np + b_np
print('c_np-->',c_np)
```

```python
# workaround to handle addition of two python lists

# add two Python lists
a_py = [1, 2]
b_py = [3, 4]

def elemwise_add(v, w):
    return [v_i + w_i for v_i, w_i in zip(v, w)]

c_py = elemwise_add(a_py,b_py)
print('c_py-->',c_py)
```
