# **Load "Raw" California Housing dataset**

```python
# California Housing dataset
import urllib.request
urllib.request.urlretrieve("https://raw.githubusercontent.com/ageron/handson-ml2/master/datasets/housing/housing.csv", "housing.csv")

import pandas as pd

# load data from csv file
housing = pd.read_csv('housing.csv')

# display the top rows
housing.head(10)
```

# Show the categories of the column "ocean_proximity"

```python
housing_cat = housing[["ocean_proximity"]]
print(type(housing_cat))

# display the count of each unique category
housing_cat["ocean_proximity"].value_counts()
```

# Get "ordinary" label encoding

```python
from sklearn.preprocessing import OrdinalEncoder

ordinal_encoder = OrdinalEncoder()
housing_cat_encoded = ordinal_encoder.fit_transform(housing_cat)
print(type(housing_cat_encoded))


import random
housing_cat_encoded[random.sample(range(housing_cat_encoded.shape[0]), 10)]
```

```python
ordinal_encoder.categories_
```

```python
import numpy as np

unique_elements, counts_elements = np.unique(housing_cat_encoded, return_counts=True)
print(np.asarray((unique_elements, counts_elements)).astype(int))
```
