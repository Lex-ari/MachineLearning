# **Load "Raw" California Housing dataset**

```python
# California Housing dataset
import urllib.request
urllib.request.urlretrieve("https://raw.githubusercontent.com/ageron/handson-ml2/master/datasets/housing/housing.csv", "housing.csv")

import pandas as pd

# load data from csv file
housing = pd.read_csv('housing.csv')

# display the top rows
housing.head(10)
```

```python
housing.info()
```

# Option 1: fill missing values using pandas



---



```python
data1 = housing.dropna(subset=["total_bedrooms"], axis = 0, inplace=False)
data1.reset_index(drop=True, inplace=True)
data1.info()
```



---



```python
data2 = housing.drop("total_bedrooms", axis = 1, inplace=False)
data2.info()
```



---



```python
median = housing["total_bedrooms"].median()

values = {"total_bedrooms": median}
data3 = housing.fillna(value=values)
data3.info()
```

# Option 2: fill missing values using sklearn

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy = "median")

housing_num = housing.drop("ocean_proximity", axis=1)
imputer.fit(housing_num)

print(imputer.statistics_)

data = imputer.transform(housing_num)
print(type(data))
print(data)
```
